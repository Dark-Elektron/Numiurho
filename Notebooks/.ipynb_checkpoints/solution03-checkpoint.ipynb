# Question 1

Represent the numbers $a = (0.1)_{10}, b = (0.2)_{10}$ and $c = (0.3)_{10}$ as elements of $F(2, 5, 2)$ and calculate $a+b$. 

Repeat the same in $F(3, 5, 2)$

## Solution

$F(\underbrace{p}_{\text{base}}, \underbrace{l}_{\text{mantissa}}, \underbrace{m}_{\text{exponent}}) \rightarrow (0.\underbrace{\alpha_1}_{\alpha_1\ne0}\alpha_2\cdots\alpha_L)p^{\beta_1, \cdots, \beta_m}$

$F(2, 5, 2) = \pm0.\alpha_1\alpha_2 \alpha_3 \alpha_4 \alpha_5 \times 2^{\pm\beta_1 \beta_2}$

Convert $(0.1)_{10}$ to binary

| | |
| :---| :----|
| $0.1 \times 2 = 0.2$ | $0$ |
| $0.2 \times 2 = 0.4$ | $0$ |
| $0.4 \times 2 = 0.8$ | $0$ |
| $0.8 \times 2 = 1.6$ | $1$ |
| $0.6 \times 2 = 1.2$ | $1$ |
| $0.2 \times 2 = 0.4$ | $0$ |
| $0.4 \times 2 = 0.8$ | $0$ |
| $0.8 \times 2 = 1.6$ | $1$ |
| $0.6 \times 2 = 1.2$ | $1$ |

$\rightarrow a = (0.1)_{10} = (0.0001100110011\cdots)_2 \approx (0.11001)_2\times 2^{-\overbrace{(11)_2}^{\text{in binary}}}$

Similarly, $b = (0.2)_{10} = (0.0011001100110\cdots)_2 \approx (0.11001)_2\times 2^{-(10)_2}$

Also, $c = (0.3)_{10} = (0.01001100110011\cdots)_2 \approx (0.10011)_2\times 2^{-(01)_2}$

$a+b = (0.11001)_2\times 2^{-(11)_2} + (0.11001)_2\times 2^{-(10)_2} = (0.00011001)_2 + (0.00110010)_2 = (0.01001011)_2 = (0.10010)_2 \times 2^{-(01)_2}$

$= (0.28125)_{10}$

Note that here, $a+b\ne c$ because of the loss of accuracy. 

#### Repeating in $F(3, 5, 2)$



| | |
| :---| :----|
| $0.1 \times 3 = 0.3$ | $0$ |
| $0.3 \times 3 = 0.9$ | $0$ |
| $0.9 \times 3 = 2.7$ | $2$ |
| $0.7 \times 3 = 2.1$ | $2$ |
| $0.1 \times 3 = 0.3$ | $0$ |
| $\vdots$ | $\vdots$ |

$\rightarrow a = (0.1)_{10} = (0.002200220022\cdots)_3 \approx (0.22002)_3\times 3^{-(02)_3}$

Similarly, $b = (0.2)_{10} = (0.0121012101\cdots)_3 \approx (0.12101)_2\times 3^{-(01)_3}$

Similarly, $c = (0.3)_{10} = (0.022002200\cdots)_3 \approx (0.22002)_3\times 3^{-(01)_3}$

$a+b = (0.22002)_3\times 3^{-(02)_3} + (0.12101)_3\times 3^{-(01)_3} = (0.002002)_3 + (0.012101)_3 = (0.021110)_3 = (0.21110)_3 \times 3^{-(01)_3}$

$= (0.27572)_{10}$

# Question 2

Study the function $f(x) = \sin(x)$ and its approximation $\tilde f(x) = -(x-\pi) + \frac{(x+\pi)^3}{6}$ at $x = \exp(1.0)$. Discuss the absolute and relative backward and forward errors for $\tilde x = 2.72$. 

Give the absolute and relative condition numbers for $f$ at the given $x$.

## Solution

We are required to evaluate the absolute forward error, absolute backward error, relative forward error, and relative backward error.

#### Forward Error

$\underbrace{\delta_y}_{\text{absolute}} = |\tilde f(\tilde x) - f(x)| =|\underbrace{\epsilon_f f(\tilde x)}_{\text{generated error}} + \underbrace{f(\tilde x) - f(x)}_{\text{propagated error}}|$

<img align="right" width="500" height="400" src="https://lh3.google.com/u/0/d/1tz2jmk1MIgYdhhIkSyNba6BpUn00w339=w1920-h902-iv2">

$\epsilon_f = \frac{\tilde f (\tilde x)-f(\tilde x)}{f(\tilde x)}$

$\underbrace{\epsilon_y}_{\text{relative}} = \frac{\delta_y}{f(x)}$


#### Backward error

$\tilde f(\tilde x) = f(\tilde x + \underbrace{\delta x)}_{\text{absolute}\\ \text{backward} \\ \text{error} }$

$\epsilon_x = \frac{|\delta_x|}{x}$

$x = e^1 = 2.718281828, \tilde x = 2.72, f(x) = \sin(x), \tilde f(x) = -(x-\pi)+\frac{(x-\pi)^3}{6}$

$\delta_y = |\tilde f(\tilde x) - f(x)| = |0.40910365 - 0.4107813| = 0.0016776$

$\epsilon_y = \frac{\delta_y}{f(x)} = \frac{0.0016776}{0.410781} = 0.0041$

$f(\underbrace{\tilde x + \delta x)}_{x^*} = \tilde f(\tilde x) \rightarrow \sin(x^*) = 0.409104 \rightarrow x^* = \pi - \arcsin(0.409104) = 2.72012112$

$\rightarrow \tilde x + \delta x = 2.72012112$

$\delta x = 0.00012112$

$\epsilon_x =\frac{0.00012112}{e^1} = 4.456\times 10^{-5}$

### Condition numbers

$\text{acond}(f) = |f'(\tilde x)| = |\cos(\tilde x)| = 0.9989$

$\text{cond}(f) = \left|\frac{x f'(\tilde x)}{f(x)}\right| = \left|\tilde x\frac{\cos(\tilde x)}{\sin(\tilde x)}\right| = \left|\tilde x\cot (\tilde x)\right| = 6.06$

# Question 3

Solve 

$$x + \ln(x) = 0$$

by a modificaiton of the (not convergent) fixed point iteration $x^{(k+1)} = \varphi(x^{(k)} = -\ln(x^{(k)})$. 

To this end, control the step size by introducing a factor $s$ according to 

$$x^{(k+1)} = x^{(k)} + s\left(\varphi(x^{(k)} - x^{(k)}\right)$$

## Solution

$x^{(k+1)} = -\ln(x^{(k)})$ is not convergent because $\varphi'(x^*) > 1$. [Not contracting]

It can be made convergent by:

#### Method 1 (Controlling the step size)
$x^{(k+1)} = x^{(k)} + s\left(\varphi(x^{(k)}) - x^{(k)}\right) \xrightarrow{x=0.5} x^{(k+1)} = x^{(k)} + 0.5\left(\varphi(x^{(k)}) - x^{(k)}\right) = \phi (x)$

|    $$ k $$               | $$0$$       | $$1$$     |$$2$$|$$\cdots$$|$$14$$|
| :-----                | :-      | :--   | :--     | :----    | :---    |
|$$x^{(k)}$$               | $$1$$       | $$0.5$$   | $$0.59657$$ | $$\cdots$$ | $$0.56714$$ |
|$$-\ln(x^{(k)})$$          | $$0$$       |$$0.69315$$| $$0.51655$$ | $$\cdots$$ | $$0.56714$$ |
|$$\phi(x^{(k)})$$      | $$0.5$$     |$$0.59657$$| $$0.55656$$ | $$\cdots$$ | $$0.56714$$ |
|$$|-\ln(x^{(k)}) - x^{(k)}|$$| $$1$$       |$$0.19315$$| $$0.08002$$ | $$\cdots$$ | $$7.4 \times 10^{-7}$$|

##### Consider

What is the best value for $s$ in this example? $s=0.36189$. Let's see why.

$\phi = x - sx - s\ln(x) \rightarrow \varphi'(x^*)=0 \rightarrow 1 -s - \frac{s}{x^*} = 0 \rightarrow s = \frac{1}{1+\frac{1}{x^*}} = 0.36189$.

The problem, however, is that we don't know this value in advance!

```python

x = symbols("x")
s = symbols("s")

f1 = x + s*(-ln(x) - x)

z = np.linspace(0, 5, 100)
s_ = 0.5
f1points = [f1.evalf(subs={x: i, s: s_}) for i in z]

# create figure
fig, ax = plt.subplots(1, 1)
fig.set_figheight(6)
fig.set_figwidth(6)

# some display settings
ax.set_xlim(-1.5, 5)
ax.set_ylim(-1.5, 5)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# plotting
ax.plot(z, f1points, 'g')
ax.plot(z, z, 'b')

# define function to visualise convergence
def conv(f, x0, s0, n_it=5):
    # n_it is the number of iterations
    
    # get x and y coordinates
    xx, yy = [x0], [0]
    for i in range(n_it):
        if i % 2 == 0:
            # change y
            # evaluate function
            y = f.evalf(subs={x: xx[i], s: s0})

            xx.append(xx[-1])
            yy.append(y)

        else:
            # change x
            xx.append(yy[-1])
            yy.append(yy[-1])
    
    print(f's: {s0}, |error|: {abs(0.56714 - xx[-1])}')
    return [xx, yy]

# #################################
# # uncomment to plot convergence
# # get convergence
# x0 = 1.5
# f1_conv = conv1(f1, x0, s_)

# # plot convergence
# ax.plot(f1_conv[0], f1_conv[1], 'go--')
# ##################################


################################

# widgets for interaction    
x0 = widgets.FloatSlider(min=0, max=5, value=0.5, step=.001)
s0 = widgets.FloatSlider(min=0, max=1, value=0.5, step=.001)

# uncomment to interact with convergence
@interact
def fixed_point(x0=x0, s0=s0):
    # remove all lines and collections
    for artist in plt.gca().lines + plt.gca().collections:
        artist.remove()
        
    # recalculate f1points
    f1points = [f1.evalf(subs={x: i, s: s0}) for i in z]
    
    # plotting
    ax.plot(z, f1points, 'r', label="$x_{k+1} = x + s(-\ln(x) - x)$")
    ax.plot(z, z, 'b')
    ax.legend()
    
    # plot convergence
    f1_conv = conv(f1, x0, s0)

    # plot convergence
    p1 = ax.plot(f1_conv[0], f1_conv[1], 'go--')
    
##################################
```

#### Method 2 (Transform equation)

$x + \ln(x) = 0 \rightarrow \ln(x) = -x \rightarrow x = e^{-x} \rightarrow x^{(k+1)} = e^{-x^{(k)}}$ which is convergent because $\varphi'(x^*) < 1$

|$$|e^{-(x^{(k)})} - x^{(k)}|$$| $$0.6321$$       |$$0.3243$$| $$0.1917$$ | $$\cdots$$ | $$7.1 \times 10^{-7}$$|

As with the last example of the fixed-point equation treated, the two formulations could be seen and interacted with below:

# Question 4

By the same technique, make Newton's method for $\sin(x) =\frac{x}{2}$ convergent when starting from $x^{(0)}=1.1$.

## Solution

Transforming $\sin(x) =\frac{x}{2}$ to the Newton method form, we get

$f(x) = x - 2\sin(x) = 0$

$f'(x) = 1 -2\cos(x)$

$x^{(k+1)} = x^{(k)} - \frac{x^{(k)} - 2\sin(x^{(k)})}{1 -2\cos(x^{(k)})}$

Let's see if this converges starting at $x^{(0)}= 1.1$

```python
x = symbols("x")

f1 = x - (x - 2*sin(x))/(1-2*cos(x))

z = np.linspace(-5, 5, 100)
f1points = [f1.evalf(subs={x: i}) for i in z]

# create figure
fig, ax = plt.subplots(1, 1)
fig.set_figheight(6)
fig.set_figwidth(9)

# some display settings
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# plotting
ax.plot(z, f1points, 'g')
ax.plot(z, z, 'b')

# define function to visualise convergence
def conv(f, x0, n_it=20):
    # get x and y coordinates
    xx, yy = [x0], [0]
    for i in range(n_it):
        if i % 2 == 0:
            # change y
            # evaluate function
            y = f.evalf(subs={x: xx[i]})

            xx.append(xx[-1])
            yy.append(y)

        else:
            # change x
            xx.append(yy[-1])
            yy.append(yy[-1])
            
    return [xx, yy]

##################################
# # uncomment to plot convergence
# # get convergence
# x0 = 1.5
# f1_conv = conv3(f1, x0)

# # plot convergence
# ax.plot(f1_conv[0], f1_conv[1], 'go--')
##################################

##################################
# uncomment to interact with convergence
def interact(x0):
    # option 2, remove all lines and collections
    for artist in plt.gca().lines + plt.gca().collections:
        artist.remove()
    
    # plotting
    ax.plot(z, f1points, 'g', label=r"$x_{k+1} = x_{k} -\frac{x_{k} - 2sin(x_{k})}{1-2cos(x_{k})}$")
    ax.plot(z, z, 'b')
    ax.legend()
    
    # plot convergence
    f1_conv = conv(f1, x0)

    # plot convergence
    p1 = ax.plot(f1_conv[0], f1_conv[1], 'go--')
    
widgets.interact(interact, x0=(-5, 5, .001))
##################################
```

# Question 5

Solve the nonlinear system of equations which defines the itersection of the unit circle and the parabola given by $3(y+0.1)^2-2 = x$. 

How many (real) solutions are there?

Choose a suitable method bearing in mind cost, accuracy and sensitivity to the initial guess.

## Solution

A unit circle is given by $x^2 + y^2 = 1.$

$x^2 + y^2 - 1 = 0$

$x - 3(y+0.1)^2 - 2 = 0 \rightarrow x= 3(y+0.1)^2 - 2$

#### What method should we use? Fixed point iteration? Newton's method?

#### We look at them both and decide the winner.

### Fixed point formulation

The basic ideas applied in the single variable case is easily generalised to multivariable problems.

1. Rearrange the equations to the fixed point formulation form
2. Determine if the form so obtained satisfies the conditions of self-mapping and contraction.
3. Solve the equation (In this case, system of equations).

#### 1. Rearranging

$x_{k+1} = \sqrt{1 - y_k^2}$

$y_{k+1} = \sqrt{\frac{x+2}{3}} - 0.1$

$\begin{equation*}
\begin{pmatrix}
x_{k+1} \\
y_{k+1} \\
\end{pmatrix} =
\begin{pmatrix}
\sqrt{1 - y_k^2} \\
\sqrt{\frac{x+2}{3}} - 0.1 \\
\end{pmatrix}
\end{equation*}$

##### Could it have been formulated some other way? You bet! But why was it not?

#### 2. Differentiate (Find the Jacobian)

$\begin{equation*}
J = 
\begin{pmatrix}
\frac{\partial f_0}{x} & \frac{\partial f_0}{y} \\
\frac{\partial f_1}{x} & \frac{\partial f_1}{y} 
\end{pmatrix} = 
\begin{pmatrix}
0 & \frac{-y_k}{\sqrt{1 - y_k^2}} \\
\frac{1}{2}\sqrt{\left(\frac{3}{x_k+2}\right)} & 0 
\end{pmatrix}
\end{equation*}$

Taking the determinant of the Jacobian and substituting the values $(x_0, y_0) = (0, 0)$

$|J| = 0$

We also take a suitable interval $x, y \in (-1, 1)$. This range is easily arrived at by looking at the fixed point formulation and the Jacobian. In this interval, we observe that the formulation is contracting and also self-mapping.

#### 3. Solve the iteration


$x_{k+1} = \sqrt{1 - y_k^2}$

$y_{k+1} = \sqrt{\frac{x+2}{3}} - 0.1$

$\begin{equation*}
\begin{pmatrix}
x_{k+1} \\
y_{k+1} \\
\end{pmatrix} =
\begin{pmatrix}
\sqrt{1 - y_k^2} \\
\sqrt{\frac{x+2}{3}} - 0.1 \\
\end{pmatrix}
\end{equation*}$

```python
x, y = symbols("x y")
f0 = (1-y**2)**0.5
f1 = ((x+2)/3)**0.5 - 0.1

xn, yn = -1, 0 # <- Initial guess can be changed here
n = 10 # number of iterations

XN, YN = [xn], [yn]
for i in range(n):
    xn = f0.subs([(x, xn), (y, yn)])
    XN.append(xn)
    yn = f1.subs([(x, XN[i]), (y, yn)])
    YN.append(yn)
    
print(f"x{n}: ", xn, f"y{n}: ", yn)
```


### Newton method

The problem could be solved either by solving the system of nonlinear equations directly or by combining both equations.

#### Method 1 (Solve the system of non-linear equations)

The basic ideas applied in the single variable case is easily generalised to multivariable problems.

1. Rearrange the equations to a homogenous form, i.e. one side equal to zero
2. Solve the equation (In this case, system of equations).

$f_0 = x^2 + y^2 - 1 = 0$

$f_1 = x - 3(y+0.1)^2 + 2 = 0$

$\begin{equation*}
\begin{pmatrix}
f_0 \\
f_1 \\
\end{pmatrix} =
\begin{pmatrix}
x^2 + y^2 - 1 \\
x - 3(y+0.1)^2 + 2 \\
\end{pmatrix}
\end{equation*}$


Newton's method for a system of equations $f_1, f_2, f_3, ...$ is given as:

$X_{n+1} = X_n - J^{-1}(X) F(X)$


$\begin{equation}
\begin{pmatrix}
x_{n+1} \\
y_{n+1}
\end{pmatrix} = 
\begin{pmatrix}
x_{n} \\
y_{n}
\end{pmatrix} - J^{-1}(x_n,y_n)
\begin{pmatrix}
f_0(x_n,y_n) \\
f_1(x_n,y_n)
\end{pmatrix}
\end{equation}$
  
The Jacobian is calculated to be

$\begin{equation*}
J(x,y) = 
\begin{pmatrix}
2x & 2y \\
1 & - 6(y+0.1) \\
\end{pmatrix}
\end{equation*}$


The inverse of the Jacobian is computed to be:

$\begin{equation*}
J^{-1}(x,y) = \frac{1}{2x(- 6(y+0.1)) - 2y}
\begin{pmatrix}
- 6(y+0.1) & -2y \\
-1 & 2x \\
\end{pmatrix}
\end{equation*}$

```python
x, y = symbols("x y")
f0 = x**2 + y**2 - 1
f1 = x - 3*(y+0.1)**2 + 2

J = Matrix([[f0.diff(x), f0.diff(y)], [f1.diff(x), f1.diff(y)]])
print("Jacobian: \n", J)
J_inv = J.inv()
print()
print("Jacobian inverse: ", J_inv)

# iterative loop
no_it = 7 # number of iterations
x0, y0 = 0.1, 0.1
X_0 = [x0, y0]
XX = [X_0]
print()
conv = [[], []]
for n in range(no_it):
    x_n, y_n = XX[n][0], XX[n][1]
    conv[0].append(x_n)
    conv[1].append(y_n)
    
    print(f"(x{n}, y{n}) = ({x_n}, {y_n})")
    
    # substitute values into expressions
    FS = [f0.subs([(x, x_n), (y, y_n)]), f1.subs([(x, x_n), (y, y_n)])]
    J_invS = J_inv.subs([(x, x_n), (y, y_n)])
    
    # convert to numpy arrays
    J_invS = np.array(J_invS.tolist())
    FS = np.array(FS)
    
    # newton step
    XX.append(XX[n] - J_invS.dot(FS))

print()
plt.close()

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figheight(4.5)
fig.set_figwidth(9)

ax1.plot(conv[0], marker='o')
ax1.set_title("Convergence Plot x")
ax1.set_xlabel("Iteration step")
ax1.set_ylabel("x")

ax2.plot(conv[1], marker='o')
ax2.set_title("Convergence Plot y")
ax2.set_xlabel("Iteration step")
ax2.set_ylabel("y")
```
